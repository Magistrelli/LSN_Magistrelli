### Esercizio 06.1
Si vuole studiare un *modello di Ising* 1D a $N=50$ gradi di libertà di spin, $\{s_i\}_{i=1,N}$. Si assume che questi possano assumere solo i valori $\pm 1$ (in unità di magnetone di Bohr $\mu_B$) e si studia il sistema in *periodic boundary condition* (sistema di bulk, $s_{N+1}=s_1$).\
L'obbiettivo è quello di stimare i valori di aspettazione delle principali grandezze termodinamiche che caratterizzano lo stato macroscopico del reticolo di spin in esame. In particolare si vuole studiare l'andamento in funzione della temperatura $T$ del sistema dell'energia interna media per particella $U/N$, della capacità termica media per particella $C/N$, della suscettività magnetica $\chi$ e della magnetizzazione media per particella $M/N$. Le prime tre grandezze devono essere studiate a campo esterno applicato nullo, $h=0$, la magnetizzazione ad $h=0.2$ (le librerie sono comunque scritte tutte considerando il caso di $h$ generico).\
Il tutto deve essere fatto attraverso delle simulazioni Monte Carlo, e si devono confrontare i risultati ottenuti con l'*algoritmo di Metropolis* con quelli raccolti col *Gibbs sampling*.

Supponendo la presenza di interazioni solo tra spin primi vicini (*near neighbours*, $n.n.$) e scegliendo le unità di misura in modo che $\mu_B=1$ and $k_B=1$, l'*Hamiltoniana* del sistema può essere scritta come
$$H = -J\sum_{i=1}^N s_i s_{i+1} - \frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1})$$
con $J$ intensità delle interazini tra primi vicini; per $J>0$ tutti gli spin tendono, a basse temperature, ad allinearsi nella stessa direzione e il materiale descritto è un ferromagnete, mentre per $J<0$ si ha un anti-ferromagnete con spin alternati quando $T \to 0$.


Il sistema considerato costituisce un *ensamble canonico*, e cioè è mantenuto a numero di particelle $N$ e temperatura $T$ (e volume) fissati. La densità di probabilità per l'occupazione degli stati possibili è quindi data dalla distribuzione di Boltzmann,

$$ \rho (\mu) = \frac{e^{-H_\mu/T}}{Z}$$

con $Z=\sum_{\mu} \exp \left( -\beta H_\mu\right)$ *funzione di partizione* e $\mu \equiv (s_1,\dots,s_N)$ particolare configurazione di spin. Per la stima dei valori di aspettazzione delle grandezze di interesse sarà quindi questa la probabilità $\rho$ che deve essere campionata dalle catene di Markov prodotte in fase di simulazione.

In questo contesto è facile non solo calcolare, ma anche campionare direttamente la probabilità condizionata che un particolare grado di libertà $s_k$ assuma un certo valore $s_k'$ fissati i valori di tutti gli altri $s_j$. Si può infatti dimostrare che si ha, già normalizzata,

$$ p_{1|(N-1)}\left(s_k'=\pm 1\ |\ {\{ s_j\}_{j \neq k}}\right) =
\frac{1}{ 1 + exp\{\mp\ 2 \beta \left[ J \left(s_{k-1}+s_{k+1} \right) + h \right] \}}$$

Ecco quindi che il sistema studiato fornisce un'ottima opportunità di testare il campionamento alla Gibbs, per la cui applicabilità è necessaria la capacità di campionare direttamente la condizionata.

In fase di simulazione ho, per entrambi gli algoritmi di campionamento, simulato dapprima $N_{burn}=5\times10^3$ step di equilibrazione, per poi effettuare altre $N_{step}=5\times10^4$ mosse per ognuno degli $N_{blk}=20$ blocchi utilizzati per il *blocking method*. Il numero di step per blocco è stato scelto in modo da avere dei raggruppamenti sufficientemente estesi perché le medie di blocco fossero tra loro indipendenti, e quindi il CLT applicabile. In questo modo si possono trovare i valori di aspettazione delle grandezze termodinamiche di interesse mediando, per ogni $T$, quelli dei singoli blocchi e valutarne l'incertezza statistica calcolandone la deviazione standard della media. 

Nei due casi una singola mossa consiste nel ripetere per $N=50$ volte la generazione di un numero intero casuale $k \in [0,N-1]$ e nel tentativo di cambiare (Metropolis), o nell'estrarre casualmente secondo $p_{1|(N-1)}(s_k'|{\{ s_j\}_{j \neq k}})$ (Gibbs), il valore del corrispondente $s_k$.\
La probabilità di transizione di prova $T(\nu|\mu)$ del Metropolis viene quindi definita implicitamente, provando, in sequenza per ogni $s_k$ estratto, ad invertirne il valore, mentre per ognuna di queste sottomosse $A(\mu' | \mu)=\min \left[ 1,q(\mu' | \mu)\right]$, con ($T(\nu|\mu)$ simmetrica)

$$ q(\mu' | \mu) = e^{-\beta \left(E_{\mu'} - E_{\mu}\right)} \ ,
\qquad \left(E_{\mu'} - E_{\mu}\right) = 2 s_k^{\mu} \left[ J \left(s_{k-1}^{\mu} + s_{k+1}^{\mu} \right) +h \right]$$

è ottentuta a partire dalla distribuzione di Boltzmann e dall'espressione di $H$; qui $\mu' \equiv (s_1,\dots,-s_k,\dots,s_N)$ e $\mu \equiv (s_1,\dots,s_k,\dots,s_N)$.\
La mossa di Gibbs invece è sempre accettata. In effetti questo può essere visto come un caso particolare dell'algoritmo di Metropolis; costruendo la probabilità di transizione di prova $T(\nu|\mu)$ nel modo descritto sopra, cambiando i valori degli $s_k$ con la probabilità condizionata, si può verificare che deve sempre essere $A=1$. 

In ogni blocco i valori di aspettazione delle grandezze di interesse sono stati calcolati attraverso le relazioni

$$ U(N,T,h) = \langle H \rangle $$

$$ C(N,T,h) = \frac{\partial U(N,T)}{\partial T} = \beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right) $$

$$ M(N,T,h) = k_B T \frac{\partial \ln Z}{\partial h} =
\left\langle \sum_{i=1}^N s_i \right\rangle $$

$$ \chi(N,T,h) = \frac{\partial M}{\partial h} =
\beta \left[ \left\langle \left( \sum_{i=1}^N s_i \right)^2 \right\rangle - \left\langle \sum_{i=1}^N s_i \right\rangle^2 \right] $$

```python
import numpy as np
from numpy import tanh, cosh, sinh, exp, sqrt
import matplotlib.pyplot as pl
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes    # zoomed up image libraries
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

In una sola dimensione il modello di Ising è risolubile anche analiticamente, e questo permette di avere dei risultati teorici di riferimento. Da $H$ si può ricavare esplicitamente

$$ Z = \lambda_1^N + \lambda_2^N $$

con $\lambda_{1,2} = e^{\beta J} \cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$. Da questa espressione si può ottenere l'energia libera di Helmholtz $F = - T \ln Z$, e infine, con opportune derivate parziali di $Z$ ed $F$, arrivare alle espressioni esplicite delle grandezze termodinamiche di interesse:

$$ \frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \cdot \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} $$
$$ \frac{C(N,T)}{N} = \frac{1}{N} \frac{\partial U(N,T)}{\partial T}
\underset{h=0}= (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} - N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}$$
$$ \frac{M(N,T,h)}{N} = -\frac{1}{N} \frac{\partial F}{\partial h} = 
\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]$$
$$ \chi(N,T) = \frac{\partial M}{\partial h}
\underset{h=0}= \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)$$

dove $U/N$, $C/N$ e $\chi$ sono calcolate a campo esterno $h=0$, mentre l'espressione per $M/N$ vale per $h$ generico, e sarà calcolata per $h=0.2$. Per campo nullo quest'ultimo risultato sarebbe infatti banalmente $M(T)/N=0 \ \forall T$: a qualunque temperatura $T$ si avrà un risultato medio nullo, in quanto non vi è alcuna direzione privilegiata per l'orientamento degli spin. Solo un campo esterno non nullo può rompere tale degenerazione.

```python
J,Ns = 1.0,50        #J, number of spins
h=0.02 #external field

#generating theorecal curves
points=100
Tlin = np.linspace(0.2,3.0,num=points)
b = 1/Tlin
th = tanh(J/Tlin)
thN= th**Ns
ch = 1/th
#predicted energy
ene = -J*( th + ch*thN )/( 1 + thN )
#predicted heat capacity
heat=((b*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
#compute predicted magnetisation (h!=0)
l1 = exp(b*J)*cosh(b*h)+sqrt(exp(2*b*J)*cosh(b*h)*cosh(b*h)-2*sinh(2*b*J))
l2 = exp(b*J)*cosh(b*h)-sqrt(exp(2*b*J)*cosh(b*h)*cosh(b*h)-2*sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (exp(b*J)*sinh(b*h)*((l1**(Ns-1))*(1+exp(b*J)*cosh(b*h)/sqrt(exp(2*b*J)*cosh(b*h)*cosh(b*h)-2*sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-exp(b*J)*cosh(b*h)/sqrt(exp(2*b*J)*cosh(b*h)*cosh(b*h)-2*sinh(2*b*J)))))/(Z)
#predicted susceptibility
X = b*exp(2*b*J)*(1-thN)/(1+thN)

obs_teo=(ene,heat,X,M)
title_obs=('Internal energy per particle','Specific heat per particle','Magnetic suscettivity','Magnetization per particle')
```

Il valore scelto per l'intensità dell'interazione a coppie è $J=1$, il range di temperature indagato è $T \in [0.5,2]$.

Le simulazioni partono dallo stato $T=\infty$, in cui ad ogni grado di libertà è assegnato un valore di spin scelto casualmente tra i due valori possibili. Per mostrare che l'$N_{burn}$ scelto è sufficiente per portare a termine la fase di equilibrazione ho rappresentato, nel grafico sottostante, l'andamento dei valori puntuali dell'energia interna totale $U$ alle temperature $T_1 = 0.5$, $T_1 = 0.95$ e $T_3 = 2$. I casi riportati sono quelli per $h=0$.

```python
T=(0.5,0.95,2.)
eteo=[-J*( (tanh(J/t)) + (1/(tanh(J/t)))*((tanh(J/t))**Ns) )/( 1 + ((tanh(J/t))**Ns) )*Ns for t in T]
title=('First 200 points','last 500 points'); labteo=['expected <U>','','']

def graph_burn(ifMetro):
    if ifMetro: name="metro"
    else: name="gibbs"

    nburn=np.genfromtxt('results/burnIn.'+name+'_0.out',max_rows=1)
    burn0=np.genfromtxt('results/burnIn.'+name+'_0.out',skip_header=1,names='x,u')
    burn10=np.genfromtxt('results/burnIn.'+name+'_6.out',skip_header=1,names='x,u')
    burn20=np.genfromtxt('results/burnIn.'+name+'_20.out',skip_header=1,names='x,u')
    burn=(burn0,burn10,burn20); labT=['T='+str(temp) for temp in T]

    pl.figure(figsize=(15,4))
    for i in range(2):
        pl.subplot(1,2,i+1)
        for j in range(3): pl.plot(burn[j]['x'],burn[j]['u'],label=labT[j])
        for j in range(3): pl.axhline(y=eteo[j],color='k',linewidth=0.8,label=labteo[j])
        if i==0: pl.xlim(-1,200)
        elif i==1: pl.xlim(nburn-500,nburn)
        pl.legend(loc='best')
        pl.xlabel('$N_{burn}$')
        if i==0: pl.ylabel('$U$',labelpad=15)
        pl.title(title[i])
        pl.grid(True)
    pl.suptitle('Burn in, '+name+' sampling')
    pl.show()
```

```python
graph_burn(True)
graph_burn(False)
```

Si vede che entrambi gli algoritmi raggiungono la condizione di stazionarietà (e quindi di corretto campionamento della distribuzione di Boltzmann) già nelle prime decine di passi, in tutto il range di temperatura.

Si può anche intuire il tipo di difficoltà che nascerebbe a campionare temperature più basse. Il fatto è che non si possono adattare gli algoritmi alle diverse temperature a cui viene studiato il sistema.\
La $T(\nu|\mu)$ del Metropolis è definita implicitamente e non ha alcun parametro libero, la mossa tentata è sempre il cambiamento di segno sequenziale di una serie casuale di spin. Di conseguenza le mosse hanno, al diminuire di $T$, un'accettazione sempre più bassa che non può essere regolata. Le configurazioni diventano sempre più correlate, in quanto quelle a bassa energia tendono a ripetersi sempre di più (si vede molto bene nel grafico del burn-in del Metropolis), e quindi per ottenere dei risultati affidabili servono blocchi sempre più estesi e dunque simulazioni sempre più lunghe.\
Anche nel caso dell'algortimo di Gibbs è evidente una sempre maggior stagnazione delle configurazioni al diminuire di $T$, dovuta qui al sempre maggior sbilanciamento di $p_{1|(N-1)}\left(s_k'=\pm 1\ |\ {\{ s_j\}_{j \neq k}}\right)$ nei due casi. In altre parole la condizionata si sbilancia sempre di più verso uno solo dei due valori accessibili. Le considerazioni sulla stima dei risultati sono quindi le stesse.

Seguono i grafici che riportano le stime delle grandezze con le loro incertezze. I due metodi forniscono risultati del tutto compatibili tra loro e con le previsioni teoriche.

```python
def obs_graph(obs_name):
    if obs_name=='ene': obs_ind=0; ylab='U'
    elif obs_name=='heat': obs_ind=1; ylab='C' 
    elif obs_name=='chi': obs_ind=2; ylab='\chi'
    elif obs_name=='mag': obs_ind=3; ylab='M'
    
    namefolder='results'
    if obs_name=='mag': namefolder+='_h'   #for mag we need h!=0
    metro=np.genfromtxt(namefolder+'/output.metro.'+obs_name+'.0',names='T,res,err')
    gibbs=np.genfromtxt(namefolder+'/output.gibbs.'+obs_name+'.0',names='T,res,err')
    obs=(metro,gibbs); label=('Metropolis','Gibbs')
    
    fig,ax=pl.subplots(figsize=(12,6))
    for i in range(2): pl.errorbar(obs[i]['T'],obs[i]['res'],yerr=obs[i]['err'],label=label[i])
    pl.plot(Tlin,obs_teo[obs_ind],color='k',linewidth=0.8,label='expected $<'+ylab+'/N>$')
    pl.xlim(0.4,2.1)
    if obs_name=="chi": pl.ylim(-10,140)
    pl.legend(loc='best')
    pl.xlabel('$T$')
    pl.ylabel('$'+ylab+'/N\ (T)$',labelpad=15)
    pl.title(title_obs[obs_ind])
    pl.grid(True)
    
    if obs_name=="heat":    #zoomed part for heat
        axins=zoomed_inset_axes(ax,2.8,loc=8)   # axes instance zoomed, zoom and position
        for i in range(2): axins.errorbar(obs[i]['T'],obs[i]['res'],yerr=obs[i]['err'])
        axins.plot(Tlin,obs_teo[obs_ind],color='k',linewidth=0.8)
        axins.set_xlim(0.52,0.92)
        axins.set_ylim(0.38,0.46)
        pl.grid(True)
        pl.yticks(visible=False)
        pl.xticks(visible=False)
        mark_inset(ax,axins,loc1=1,loc2=3,fc='none',ec='0.6')
    
    pl.show()
```

```python
obs_graph("ene")
```

Al diminuire di $T$ si riduce anche $U/N$: l'agitazione termica diminuisce e gli spin tendono ad essere tutti allineati (materiale ferromagnetico), e quindi $U/N \to -1$ per $T \to 0$.

```python
obs_graph("heat")
```

```python
obs_graph("mag")
```

A basse $T$ la suscettività diventa molto alta per $h\to0$. Questo è conseguenza della rottura di simmetria legata all'accensione del campo esterno.
