### Esercizio 04.1

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl

nG=6              #numbers of graphs (max 8)
```

```python
pl.figure(figsize=(15,15))
step=1
while step<=nG:

    ekin=np.genfromtxt('results/output.solid.ekin.'+str(step-1),skip_header=1,names='x,e')
    epot=np.genfromtxt('results/output.solid.epot.'+str(step-1),skip_header=1,names='x,e')
    etot=np.genfromtxt('results/output.solid.etot.'+str(step-1),skip_header=1,names='x,e')
    temp=np.genfromtxt('results/output.solid.temp.'+str(step-1),skip_header=1,names='x,e')
    
    pl.subplot(3,2,step)
    pl.plot(ekin['x'],ekin['e'],label='Kinetic Energy')
    pl.plot(epot['x'],epot['e'],label='Potential Energy')
    pl.plot(etot['x'],etot['e'],label='Total Energy')
    pl.plot(temp['x'],temp['e'],label='Temperature')
    pl.axhline(y=0.8,color='k',linewidth=0.8,label='Desired T')
    pl.legend(loc='best')
    if(step>4): pl.xlabel('$N$ step')
#   pl.ylabel('$I$',labelpad=15)
    pl.title('Iteration '+str(step))
    pl.grid(True)
    
    step+=1
    
pl.suptitle('Solid State',y=0.92)
pl.show()
```

### Esercizio 04.2

```python
Ekin=np.genfromtxt('results/ave.solid.ekin.out',usecols=(0,2,3),names='x,res,err')
Epot=np.genfromtxt('results/ave.solid.epot.out',usecols=(0,2,3),names='x,res,err')
Etot=np.genfromtxt('results/ave.solid.etot.out',usecols=(0,2,3),names='x,res,err')
temp=np.genfromtxt('results/ave.solid.temp.out',usecols=(0,2,3),names='x,res,err')

pl.figure(figsize=(18,8))

pl.subplot(2,2,1)
pl.errorbar(Ekin['x'],Ekin['res'],yerr=Ekin['err'])
#pl.xlabel('$N$')
pl.ylabel('$E_{kin}$')
pl.title('Kinetic Energy')
pl.grid(True)

pl.subplot(2,2,2)
pl.errorbar(Epot['x'],Epot['res'],yerr=Epot['err'])
#pl.xlabel('$N$')
pl.ylabel('$E_{pot}$')
pl.title('Potential Energy')
pl.grid(True)

pl.subplot(2,2,3)
pl.errorbar(Etot['x'],Etot['res'],yerr=Etot['err'])
pl.xlabel('$N$ block')
pl.ylabel('$E_{tot}$')
pl.title('Total Energy')
pl.grid(True)

pl.subplot(2,2,4)
pl.errorbar(temp['x'],temp['res'],yerr=temp['err'])
pl.xlabel('$N$ block')
pl.ylabel('$T$')
pl.title('Temperature')
pl.grid(True)

pl.suptitle('Solid State')
pl.show()
```

### Esercizio 04.3

```python
kB,eps_kB=1.98064852e-23,120.       #Boltzmann constant, eps/kB
eps=eps_kB*kB;

Ekin=np.genfromtxt('results/Ar.ave.solid.ekin.out',usecols=(0,2,3),names='x,res,err')
Epot=np.genfromtxt('results/Ar.ave.solid.epot.out',usecols=(0,2,3),names='x,res,err')
Etot=np.genfromtxt('results/Ar.ave.solid.etot.out',usecols=(0,2,3),names='x,res,err')
temp=np.genfromtxt('results/Ar.ave.solid.temp.out',usecols=(0,2,3),names='x,res,err')

pl.figure(figsize=(18,8))

pl.subplot(2,2,1)
pl.errorbar(Ekin['x'],Ekin['res']*eps,yerr=Ekin['err']*eps)
#pl.xlabel('$N$')
pl.ylabel('$E_{kin} [J]$')
pl.title('Kinetic Energy')
pl.grid(True)

pl.subplot(2,2,2)
pl.errorbar(Epot['x'],Epot['res']*eps,yerr=Epot['err']*eps)
#pl.xlabel('$N$')
pl.ylabel('$E_{pot} [J]$')
pl.title('Potential Energy')
pl.grid(True)

pl.subplot(2,2,3)
pl.errorbar(Etot['x'],Etot['res']*eps,yerr=Etot['err']*eps)
pl.xlabel('$N$ block')
pl.ylabel('$E_{tot} [J]$')
pl.title('Total Energy')
pl.grid(True)

pl.subplot(2,2,4)
pl.errorbar(temp['x'],temp['res']*eps_kB,yerr=temp['err']*eps_kB)
pl.xlabel('$N$ block')
pl.ylabel('$T [K]$')
pl.title('Temperature')
pl.grid(True)

pl.suptitle('Solid State')
pl.show()
```
