### Esercizio 04.1

L'uso delle due librerie <font face="Courier">Experiment</font> e <font face="Courier">MolDyn_NVE</font> permette di simulare, tramite l'*algoritmo di Verlet*, l'evoluzione temporale di un sistema di particelle interagenti tramite un potenziale di tipo Lennard-Jones

$$V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r}\right)^{12} - \left( \frac{\sigma}{r}\right)^6 \right] $$

sottoposto a condizioni al contorno periodiche (limite termodinamico per la descrizione di un sistema di bulk).

L'algoritmo di Verlet necessita, per il calcolo delle posizioni delle particelle ad un istante di tempo successivo, non solo delle posizioni attuali, ma anche di quelle ad uno step precedente:

$$ \vec{r}(t + \delta t) = 2 \vec{r}(t) - \vec{r}(t - \delta t) + \vec{a}(t) {\delta t}^2$$

con $\vec{r}(t)$ vettore che descrive completamente la configurazione spaziale del sistema raccogliendo le posizioni al tempo $t$ di tutte le particelle e $\vec{a}(t)$ vettore che ne raccoglie le accelerazioni.

Basandosi sulla generica libreria <font face="Courier">Experiment</font>, la più specifica <font face="Courier">MolDyn_NVE</font> permette di simulare le proprietà di un sistema a numero di particelle $N$, volume (del box) $V$ ed energia $E$ fissate (*sistema microcanonico*), sia partendo solamente da una configurazione iniziale <font face="Courier">config.0</font>, sia ripartendo da una precedente simulazione e leggendo inoltre la penultima configurazione spaziale <font face="Courier">old.0</font>.\
Nel primo caso il programma genera in modo casuale le velocità iniziali $\vec{v}_{now}$ (subito imponendo una velocità del centro di massa nulla), le riscala per ottenere un valore associato di energia cinetica compatibile con la temperatura $\overline{T}$ a cui si vuole indagare il sistema, $\vec{v}(t_0) = \vec{v}_{now} \sqrt{\overline{T}/T_{now}}$, e le utilizza per ricostruire la (fittizia) configurazione $\vec{r}(t_0 - \delta t) = \vec{r}(t_0) - \vec{v}(t_0) \delta t$ necessaria per l'avvio dell'algoritmo. Qui nasce però un problema che costituisce una delle motivazioni principali per introdurre la possibilità delle ripartenze. Il fatto è che il sistema, partendo da una configurazione "artificiale" (tipicamente un reticolo cristallino cubico qualunque sia lo stato che si vuole studiare), va inizialmente incontro ad una fase di assestamento, al termine della quale viene raggiunto un equilibrio caratterizzato da una temperatura in generale diversa da quella imposta nel file di input.\
La possibilità di ripartire più volte consente di risolvere il problema. A questo scopo all'inizio di ogni ripartenza viene eseguito uno step dell'algoritmo e calcolata la velocità $\vec{v}(t_0 + \delta t/2)$ associata al passaggio da $\vec{r}(t_0)$ a $\vec{r}(t_0 + \delta t)$. Questa viene quindi riscalata per ottenerne una maggiormente compatibile con il valore di temperatura al quale si vuole studiare il sistema, e con la nuova velocità si ricalcola un $\vec{r}_{new}(t_0) = \vec{r}(t_0 + \delta t) - \vec{v}(t_0 + \delta t/2) \delta t$ più adeguato a questo valore. Da qui inizia quindi la vera e propria simulazione. Il meccanismo è automatico e permette di ottenere, dopo un numero sufficiente di ripartenze, un sistema caratterizzato all'equilibrio da una temperatura media il più vicino possibile a quella fornita dal file di input.

Per il resto l'algoritmo di Verlet non necessita della conoscenza della velocità delle particelle per simularne l'evoluzione. Questa serve però se si vogliono stimare proprietà macroscopiche del sistema quali l'energia cinetica $\langle K \rangle$, l'energia totale $\langle E_{tot} \rangle$ e la temperatura $\langle T \rangle$. Note le configurazioni $\vec{r}(t + \delta t)$ e $\vec{r}(t - \delta t)$ le velocità delle particelle possono essere stimate come

$$ \vec{v}(t) = \frac{\vec{r}(t + \delta t) - \vec{r}(t - \delta t)}{2 \delta t}$$

Da queste è possibile ottenere, in termini di unità ridotte ($V^\star = V/\epsilon$, $K^\star = K/\epsilon$, $T^\star = k_B T/\epsilon$)

$$ \langle T^\star \rangle = \frac{2 \langle K^\star \rangle}{3N}$$\
$$ \langle E_{tot}^\star \rangle = \langle K^\star \rangle + \langle E_{pot}^\star \rangle$$

dove $\langle K^\star \rangle$ è dato dalla semisomma dei quadrati delle velocità delle particelle (tutto sempre trattato in unità ridotte), $\langle E_{pot}^\star \rangle$ dalla somma di tutti i contributi di interazioni a coppie

$$V^\star(r) = 4 \left[ \left( \frac{1}{r}\right)^{12} - \left( \frac{1}{r}\right)^6 \right] $$

per $r<r_c$ e $V^\star(r)=0$ altrove, con $r_c$ raggio di cut-off del potenziale.

Ho simulato tre fasi dello stesso sistema di $N_{part}=108$ particelle, tutte con step temporali $\delta t^\star = 5 \times 10^{-4}$:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Subito sotto riporto, per tutte e tre le fasi, una serie di grafici che mostrano come, con qualche ripartenza (per ognuna $N_{step}=10^3$), si riesca ad ottenere un sistema stabilizzato su una temperatura media praticamente uguale a quella richiesta, e sul quale quindi si potrà simulare la misura delle grandezze macroscopiche di interesse nelle condizioni desiderate.\
I risultati sono qui riportati ogni $10$ step temporali; non è stato eseguito il calcolo delle incertezze (l'obbiettivo è solo quello di vedere l'iniziale stabilizzazione del sistema).

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl
```

```python
nG=6              #numbers of graphs (max 8)

def suptitle(state):
    if state=="solid": state="Solid"
    elif state=="liquid": state="Liquid"
    elif state=="gas": state="Gas"
    pl.suptitle(state+' State')
    print()

def state_equil(state,T_des):   #state is "solid", "liquid" or "gas"
    pl.figure(figsize=(16,8))
    step=1
    while step<=nG:

        ekin=np.genfromtxt('results/output.'+state+'.ekin.'+str(step-1),skip_header=1,names='x,e')
        epot=np.genfromtxt('results/output.'+state+'.epot.'+str(step-1),skip_header=1,names='x,e')
        etot=np.genfromtxt('results/output.'+state+'.etot.'+str(step-1),skip_header=1,names='x,e')
        temp=np.genfromtxt('results/output.'+state+'.temp.'+str(step-1),skip_header=1,names='x,e')
        mes=(ekin,epot,etot,temp)
        label=('Kinetic Energy','Potential Energy','Total Energy','Temperature')
    
        pl.subplot(2,3,step)
        for i in range(4): pl.plot(mes[i]['x'],mes[i]['e'],label=label[i])
        pl.axhline(y=T_des,color='k',linewidth=0.8,label='Desired T')
        pl.legend(loc='best')
        if(step>3): pl.xlabel('$N_{step}$')
        pl.title('Iteration '+str(step))
        pl.grid(True)
    
        step+=1
    
    suptitle(state)
    pl.show()
```

```python
state_equil("solid",0.8)
state_equil("liquid",1.1)
state_equil("gas",1.2)
```

Si vede facilmente che dopo $6$ ripartenze tutte le fasi del sistema raggiungono una temperatura media molto vicina a quella richiesta.

### Esercizio 04.2

Le librerie utilizzate sono anche in grado, a seguito dell'iniziale fase di equilibrazione, di misurare le grandezze termodinamiche del sistema e di fornirne una valutazione dell'errore statistico attraverso il blocking method.

Segue una serie di grafici che riportano le misurazioni effettuate per l'energia potenziale media per particella, $\langle E_{pot} \rangle/N_{part}$, per l'energia cinetica media per particella, $\langle K \rangle/N_{part}$, per l'energia totale media per particella, $\langle E_{tot} \rangle/N_{part}$ e per la temperatura media $\langle T \rangle$ del sistema in funzione del numero di blocchi analizzati.\
Per ottenerle ho innnanzitutto stabilizzato il sistema (in ogni fase) con $8$ ripartenze da $N_{step}=1000$ passi ciascuna (<font face="Courier">main01.x</font>) ed ho quindi eseguito una simulazione della sua evoluzione temporale per $N_{step}=10^4$ passi. Supponendo che il sistema sia in equilibrio durante tutta quest'ultima simulazione le misure effettuate per le grandezze macroscopiche ad ogni passo sono tra loro tutte equivalenti. Le ho suddivise in $N_{blk}=100$ blocchi e ne ho quindi studiato l'evoluzione della media e dell'incertezza statistica all'aumentare del numero di blocchi utilizzati.

```python
kB,eps_kB=1.98064852e-23,120.          #Boltzmann constant, eps/kB
eps=eps_kB*kB;

def state_meas(state,ifAr):            #print measures graphs, ifAr=True if Argon is studied
    if ifAr: whichfile = "results/Ar." #Argon, real units
    else: whichfile = "results/"       #reduced units

    Ekin=np.genfromtxt(whichfile+'ave.'+state+'.ekin.out',usecols=(0,2,3),names='x,res,err')
    Epot=np.genfromtxt(whichfile+'ave.'+state+'.epot.out',usecols=(0,2,3),names='x,res,err')
    Etot=np.genfromtxt(whichfile+'ave.'+state+'.etot.out',usecols=(0,2,3),names='x,res,err')
    temp=np.genfromtxt(whichfile+'ave.'+state+'.temp.out',usecols=(0,2,3),names='x,res,err')

    mes=[Epot,Ekin,Etot,temp]
    y_lab=['$E_{pot}$','$E_{kin}$','$E_{tot}$','$T$']
    title=('Potential Energy','Kinetic Energy','Total Energy','Temperature')
    if ifAr:                           #conversion in real unit, explicit in the labels
        for i in range(3): mes[i]['res']*=eps; mes[i]['err']*=eps; y_lab[i]+='$\ [J]$'
        temp['res']*=eps_kB; temp['err']*=eps_kB; y_lab[3]+='$\ [K]$'

    pl.figure(figsize=(18,8))
    for i in range(4):
        pl.subplot(2,2,i+1)
        pl.errorbar(mes[i]['x'],mes[i]['res'],yerr=mes[i]['err'])
        if i>1: pl.xlabel('$N_{blk}$')
        pl.ylabel(y_lab[i])
        pl.title(title[i])
        pl.grid(True)        
        
    suptitle(state)
    pl.show()
```

```python
state_meas("solid",False)
state_meas("liquid",False)
state_meas("gas",False)
```

Le temperature delle tre fasi sono molto vicine a quelle imposte nei file di input. Si possono inoltre notare le ottime proprietà di conservazione dell'energia totale da parte dell'algoritmo di Verlet anche sul lungo periodo. Nonostante le varie grandezze misurate non siano infatti self-compatibili dall'inizio alla fine dell'evoluzione, le variazioni relative di energia totale sono, in tutti i casi, molto piccole. Per ridurre le fluttuazioni temporali delle grandezze misurate si potrebbero eseguire simulazioni più lunghe aumentando sia il numero di blocchi sia, soprattutto, il numero di step per blocco. Anche aumentare il numero di ripartenze iniziali (soprattutto nel caso del gas, come sarà evidente dalla lezione07) può aiutare a migliorare i risultati.
