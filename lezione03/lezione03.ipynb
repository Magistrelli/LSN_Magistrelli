### Esercizio 03.1
L'obbiettivo qui è quello di stimare il giusto prezzo $S_{pr}$ per un'opzione *plain vanilla*, dato in generale dal valore di aspettazione del guadagno previsto $S_{gain}$ per l'holder riscalato del fattore $e^{-rT}$, cioè secondo l'inverso del rate di crescita del patrimonio investito $S_{pr}$ legato al tasso fisso garantito $r$, supposto costante, passato un tempo pari a quello di consegna $T$ dell'opzione (nel limite di riscossione continua degli interessi), $ S_{pr} = \langle S_{gain} e^{-rT} \rangle$

Il valore del guadagno, e quindi del prezzo dell'opzione, è in generale fissato, oltre che dalla scadenza $T$ e dal tasso $r$, anche dallo strike price $K$ fissato per l'asset alla scadenza, dal prezzo $S(t)$ dello stesso al momento della stipulazione del contratto e dalla sua volatilità $\sigma$, che nel seguito verrà assunta costante. I valori scelti per questi dati nell'esercizio sono riportati nello specchietto qui sotto.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl

from numpy import sqrt,log,exp
from scipy.special import erf

S0 = 100.; K = 100.; T = 1.; r = 0.1; sigma = 0.25    # data
```

Sotto le ipotesi di mercato completamente efficiente (*Markov assumption*, l'evoluzione del mercato dipende ad ogni tempo solo dal prezzo istantaneo del bene) e di prezzo dell'asset che evolve secondo un moto browniano geometrico a drift $\mu$ e volatilità $\sigma$ costanti, $S(t) \sim GBM(\mu,\sigma^2)$, il risultato corretto è fornito dalla soluzione analitica di Black-Scholes. La sua implementazione e i risultati da essa forniti nei casi di *call* (diritto di acquisto) e *put* (diritto di vendita) sono riportati subito sotto.

```python
def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

Questi, come quelli che seguiranno, sono risultati statisticamente esatti, nel senso che sotto l'ipotesi di evoluzione secondo un $GBM$ non viene commesso alcun errore di discretizzazione dell'evoluzione temporale. In altre parole la distribuzione congiunta delle sequenze di prezzi che si possono riprodurre in funzione di una serie di tempi di misura successivi è la stessa del moto brawniano geometrico continuo, indipendentemente dalla larghezza dei singoli intervalli temporali.

Quello che richiedeva l'esercizio era quindi di utilizzare la previsione teorica di Black e Scholes per verificare la validità di una previsione del prezzo ottenuta con metodi Monte Carlo. Il tutto si concretizza nella stima di $\langle e^{-rT} \max\left[0,S(T)-K\right] \rangle$ per un'opzione *call* e di $\langle e^{-rT} \max\left[0,K-S(T)\right] \rangle$ per una *put*.\
Questo è stato fatto
1. stimando direttamente il prezzo finale attraverso l'espressione carateristica per l'evoluzione di un $GBM$, $$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$ dove $W(T) \sim \mathcal{N}(0,T)$ è un processo di Wiener
2. ottenendo il prezzo finale a seguito di una suddivisione dell'intervallo $[0,T]$ in $100$ sottointervalli e di un'evoluzione $S(0)$, $S(t_1)$, $S(t_2)$, ... $S(t_{100}=T=1)$ secondo un $GBM(r,\sigma^2)$ discretizzato e descritto quindi da $$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$ con gli $Z_i$ tutti estratti casualmente da una distribuzione $\mathcal{N}(0,1)$

Il primo punto è possibile solamente grazie al fatto che l'equazione differenziale stocastica $dS(t) = \mu S(t) dt + \sigma S(t) dW(t)$ associata al moto browniano geometrico è risolubile esattamente. Per processi più complicati la discretizzazione dell'evoluzione temporale risulterebbe indispensabile. In questo caso comunque entrambi gli approcci sono esatti, e quindi dovranno essere compatibili tra loro e con la soluzione di Black e Scholes.

Di seguito i risultati. Le incertezze sono state ottenute ancora col blocking method, dividendo le $M=10^6$ differenti misurazioni del prezzo in $N=250$ blocchi.
